Para realizar la comparación entre los modelos aplicados a los diferentes conjuntos de datos, se muestra la tabla de errores ordenada de mayor a menor según el valor de R². ## **Líbrerias y modulos necesarios**

```python
import pandas as pd
```

## **Modelo 1: Wind speed 100 metros (Sensor 1)**

```python
resultados_benchmark = pd.read_pickle('resultados1.pkl')
resultados_original = pd.read_pickle('resultados1_original.pkl')
resultados_modelo1 = pd.concat([resultados_benchmark, resultados_original], ignore_index = True)
resultados_modelo1 = resultados_modelo1.sort_values(by='R²', ascending = False)
resultados_modelo1
```

## **Conclusión**

Al observar los resultados, se concluye que la aplicación del modelo XGBoost, combinado con métodos de optimización como GWO y WOA, si bien no mejora drásticamente la precisión, sí genera una mejora notable en los tiempos de ejecución en comparación con el uso de GridSearch con XGBoost. Además, los modelos optimizados con GWO y WOA superan a los demás modelos en cuanto a las métricas evaluadas y cumplen con los supuestos del modelo, lo que indica que los modelos **no lineales**, como XGBoost, son los más adecuados para este conjunto de datos. Este análisis resalta la importancia de considerar tanto la precisión como la eficiencia en los tiempos de ejecución, lo que hace que los modelos no lineales sean una opción más robusta y escalable para este tipo de problemas.

Recordemos que, XGBoost al ser un **boosting** de árboles de decisión, aprovecha la capacidad de los árboles para capturar relaciones no lineales complejas en los datos. Además, al ser un modelo basado en árboles, no está restringido a un espacio de soluciones lineales, lo que le permite ajustarse a datos más complejos y no lineales. Por lo tanto, al implementarlo con algoritmos de optimización se espera que genere resultados más óptimos y eso fue lo que se obtuvo, confirmando así que la implementación de estos modelos fue acertada.
